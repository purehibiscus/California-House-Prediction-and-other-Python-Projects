```python
from bs4 import BeautifulSoup
import requests

url = 'https://simple.wikipedia.org/wiki/List_of_European_countries'
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
# Verifying the table class
table = soup.find('table', class_='wikitable sortable references-small jquery-tablesorter')
```

```python
table = soup.find('table')
```

```python
print(table)
```

```python
# Extracting table headers
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())

# Printing table headers
print("Table Headers:")
print(headers)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = headers)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]: # Start after position 1 and loop through
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
  #  print(individual_row_data)
    length = len(df)
    df.loc[length] = individual_row_data
```
